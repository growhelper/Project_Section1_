**3) EDA와 데이터 전처리**

```python
###1차 (직관적)EDA 및 전처리###

def step1_eda(df_amend):

  #1)type object -> datetime 쓸 수 있도록 변경 , 표시형식 일치화
  from datetime import date
  from datetime import datetime
  
  Order_Date=[]
  
  for i in range(len(df_amend)):
    try:
      Order_Date.append(pd.to_datetime(df_amend['Order Date'][i],format="%d/%m/%Y").strftime('%Y-%m-%d'))
    except:
      Order_Date.append(pd.to_datetime(df_amend['Order Date'][i],format="%d-%m-%Y").strftime('%Y-%m-%d'))
      
  Ship_Date=[]
  
  for j in range(len(df_amend)):
    try:
      Ship_Date.append(pd.to_datetime(df_amend['Ship Date'][j],format="%d/%m/%Y").strftime('%Y-%m-%d')) 
    except:
      Ship_Date.append(pd.to_datetime(df_amend['Ship Date'][j],format="%d-%m-%Y").strftime('%Y-%m-%d'))
      
  
  df_amend['Order Date']=Order_Date
  df_amend['Ship Date']=Ship_Date

  #2)OrderDate / ShipDate 주말/주중 나누기-주말은 true / 주말이 아닌 경우 false
  order_week=[]
  def is_weekend(d):
    return d.weekday() > 4
    
  for k in range(len(df_amend)):
    order_week.append(is_weekend(pd.to_datetime(df_amend['Order Date'][k],format="%Y-%m-%d")))
    
  ship_week=[]
  for l in range(len(df_amend)):
    ship_week.append(is_weekend(pd.to_datetime(df_amend['Ship Date'][l],format="%Y-%m-%d")))
    
  df_amend['order_week']=order_week
  df_amend['ship_week']=ship_week


  #3)LeadTime 생성 -> ship mode와 연관성 있을지?
  LeadTime=[]
  for m in range(len(df_amend)):
    LeadTime.append(pd.to_datetime(df_amend['Ship Date'][m],format="%Y-%m-%d")-pd.to_datetime(df_amend['Order Date'][m],format="%Y-%m-%d"))
    
  df_amend['LeadTime']=LeadTime

  #4)상품정보를 쉼표(,)를 기준으로 split하여 상품명 / 규격으로 나눠서 피쳐생성
  df_amend[['ITEM', 'SPEC']] = pd.DataFrame(df_amend['Product Name'].str.split(',', 1).tolist())


  #5)표준화 할 수 있는 제조원 위해 첫번째 공백 전까지만 추출
  import re
  Maker =[]
  for o in range(len(df_amend)):
    Maker.append(re.findall(r'\S+', df_amend['ITEM'][o])[0])
    
  df_amend['Maker']=Maker


  #6)표준화 할 수 있는 상품 위해 마지막 공백 공백 이후 추출 / maybe ITEM의 끝 공백 뒷 부분을 Just_ITEM으로 추출해보기
  Just_Item=[]
  for p in range(len(df_amend)):
    Just_Item.append(df_amend['ITEM'][p].split(' ')[-1])
    
  df_amend['Just_Item']=Just_Item


  #7)판매단가/매입단가/제품매익율/판매금액/매입금액/운송비/최종매익율 설정 + order 별 상품 기준 매익율/전체 cost 기준 매익율
  df_amend['selling price']=df_amend['Sales']/df_amend['Quantity'] #제품 판매단가
  df_amend['purchase price']=(df_amend['Sales']-df_amend['Profit'])/df_amend['Quantity'] #제품 매입단가
  df_amend['Sales_Amount']=df_amend['Sales']-(df_amend['Sales']*df_amend['Discount']) #제품 판매금액 *discount 적용
  df_amend['Purchase_Amount']=df_amend['purchase price']*df_amend['Quantity'] #제품 매입금액
  df_amend['Product Margin Rate']=round(1-(df_amend['Purchase_Amount']/df_amend['Sales_Amount']),3) #제품 기준 매익율
  df_amend['Total_Cost']=df_amend['Purchase_Amount']+df_amend['Shipping Cost'] #총 cost
  df_amend['Net_Profit']=df_amend['Sales_Amount'] - df_amend['Total_Cost'] #총 순이익
  df_amend['Order Margin']=round(df_amend['Net_Profit']/df_amend['Sales_Amount'],3) # order 매익율
  
  #8)order 별 discount 영향율
  df_amend['Discount_effect']=round(-(df_amend['Sales']*df_amend['Discount'])/df_amend['Sales_Amount'],3)
  
  #9)shipping cost 영향율
  df_amend['Shipping_effect']=round(-(df_amend['Shipping Cost'])/df_amend['Sales_Amount'],3)

```

```python
#순열중요도
from sklearn.pipeline import Pipeline
# encoder, imputer를 preprocessing으로 묶었습니다. 후에 eli5 permutation 계산에 사용합니다
pipe = Pipeline([
    ('preprocessing', make_pipeline(SimpleImputer(),StandardScaler())),
    ('LGBM', LGBMClassifier(class_weight='balanced')) 
])
```

```python
pipe.fit(X_train, y_train)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import eli5
from eli5.sklearn import PermutationImportance

# permuter 정의
permuter = PermutationImportance(
    pipe.named_steps['LGBM'], # model
    scoring='precision', # metric
    n_iter=5, # 다른 random seed를 사용하여 5번 반복
    random_state=2
)

# permuter 계산은 preprocessing 된 X_val을 사용합니다.
X_val_transformed = pipe.named_steps['preprocessing'].transform(X_val)

# 실제로 fit 의미보다는 스코어를 다시 계산하는 작업입니다
permuter.fit(X_val_transformed, y_val);
```

```python
feature_names = X_val.columns.tolist()
pd.Series(permuter.feature_importances_, feature_names).sort_values()
```

```python
# 특성별 score 확인
eli5.show_weights(
    permuter, 
    top=None, # top n 지정 가능, None 일 경우 모든 특성 
    feature_names=feature_names # list 형식으로 넣어야 합니다
)
```

```python
minimum_importance = 0.001
mask = permuter.feature_importances_ > minimum_importance
features = X_train.columns[mask]
X_train_selected = X_train[features] # 유의미한 train feature만 남기기
X_val_selected = X_val[features] # 유의미한 val feature만 남기기
```

```python
#Hyperparameter tuning, etc. 최소 2개 이상의 모델을 만들어서 validation 점수를 보고

from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
# pipe_하이퍼파라미터 튜닝 ver.

pipe_pick = make_pipeline(
    SimpleImputer(),
    StandardScaler(),
    LGBMClassifier(class_weight='balanced')
)    


parameters = {
    
    'simpleimputer__strategy': ['mean', 'median'],
    'lgbmclassifier__n_estimators': (100,500),   
    'lgbmclassifier__max_bin': range(1, 15, 2),
    'lgbmclassifier__max_depth': range(1, 15, 2), 
    'lgbmclassifier__max_features': range(1, 14, 2), 
    'lgbmclassifier__learning_rate' : (0.001,0.7)
}



LGBMC_classifier = RandomizedSearchCV(
    pipe_pick, 
    param_distributions=parameters, 
    n_iter=20, 
    cv=3,
    scoring='precision',
    verbose=1,
    n_jobs=-1

)

LGBMC_classifier.fit(X_train_selected, y_train) 

#lgbm 모델이 f1을 향해서 학습해 갈때, precision으로 적당한 억제코자 함
#callbacks=[es,mc] 오류
#오류:es = EarlyStopping(monitor='precision', mode='max', verbose=1, patience=500)  early_stopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=0, mode='auto')
#오류:mc = ModelCheckpoint('best_model.h5', monitor='precision', mode='max', save_best_only=True)
```

```python
#PDP1
"""비현실적인 독립변수간의 상관관계가 없다는 가정하에 진행한다 . 최대2개 변수와 target간의 관계? interact plot
샵은 샘플 하나에 대한 피쳐 영향도들을 각각 알 수 있음. 샵벨류는 모델의 성능이 좋아야 신뢰도가 좋다 . 특성중요도와 달리 음에 대한 영향력까지 같이 적용해준다. 샘플수를 잘 조절해야 계산시간이 적게 걸린다"""

from pdpbox.pdp import pdp_interact, pdp_interact_plot

features = ['Shipping_effect', 'LeadTime']
interaction = pdp_interact(
    model=best_pipe, 
    dataset=X_test_selected,
    model_features=X_test_selected.columns, 
    features=features
)
pdp_interact_plot(interaction, plot_type='grid', 
                  feature_names=features);
```
